```python
%matplotlib qt5
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x-values')
plt.ylabel('$y = \sin(x)$ and $y = \cos(x)$')
plt.title('Plots of $\sin(x)$ and $\cos(x)$ \
on $[-\pi, \pi]$')
plt.legend(['sin(x)','cos(x)'])
plt.show()
```

```python
%matplotlib qt5
x1 = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x1)
x2 = np.linspace(-np.pi, np.pi, 10)
y2 = np.cos(x2)
fig, ax = plt.subplots()
ax.plot(x1,y1,'-',lw=5,c='blue',label='Sin(x)')
ax.plot(x2,y2,'*',ms=25,markerfacecolor='red',\
        markeredgecolor='blue',label='Cos(x)')
ax.set_xlabel('x-values')
ax.set_ylabel('$y = \sin(x)$ and $y = \cos(x)$')
ax.set_title('Plots of $\sin(x)$ and $\cos(x)$ \
on $[-\pi, \pi]$')
ax.legend(loc=4)
plt.show()
```

```python
%matplotlib inline
fig,ax = plt.subplots(2,2)
ax[0,0].plot(x1,y1)
ax[0,0].set_xlabel('x-values')
ax[0,0].set_ylabel('sin(x)')
ax[0,0].set_title('Plot of sin(x)')
ax[1,1].plot(x2,y2,'o')
ax[1,1].set_xlabel('x-values')
ax[1,1].set_ylabel('cos(x)')
ax[1,1].set_title('Plot of cos(x)')
ax[0,1].plot(x1)
ax[0,1].set_title('Plot of x=y')
ax[1,0].remove()
plt.subplots_adjust(wspace=0.2,hspace=0.3)
plt.suptitle('Plots of sin(x), cos(x) and x=y')
plt.show()
```

```python
%matplotlib qt5
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
x = np.linspace(-8,8,1000)
f = np.sin(2*np.pi+x)
g = np.cos(2*np.pi-x)

fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x,f,c='b',lw=2,label='Plot of $f(x)$')
ax.plot(x,g,c='m',lw=2,label='Plot of $g(x)$')
ax.fill_between(x,f,g,where=(f > g),color='green',\
                alpha=0.5,label='$\sin(x) > \cos(x)$')
ax.fill_between(x,f,g,where=(f < g),color='red',\
                alpha=0.5,label='$\sin(x) < \cos(x)$')
ax.legend(ncol=4, loc=3, bbox_to_anchor=(0,1))

plt.show()
```

```python
# first generated the x,y values over which the function to be evaluated.
x = y = np.linspace(-10,10,1000)

# now generate the $XY-$ grid using numpy.meshgrid
X,Y = np.meshgrid(x,y)

# now evaluate the function values on the above grid
Z = np.cos(X)*np.cos(Y)*np.exp(-(X/5)**2-(Y/5)**2)

fig, ax = plt.subplots()
h = ax.contour(X,Y,Z,50,cmap=mpl.cm.bone)
plt.colorbar(h)

fig, ax = plt.subplots()
h = ax.contourf(X,Y,Z,50,cmap=mpl.cm.bone)
plt.colorbar(h)

fig, ax = plt.subplots()
h = ax.pcolor(X,Y,Z,cmap=mpl.cm.winter,alpha=0.5)
plt.colorbar(h)
```

```python
fig, ax = plt.subplots(figsize=(6,5))
ax.remove()
ax = Axes3D(fig,auto_add_to_figure=False)
fig.add_axes(ax)
#h = ax.contour3D(X,Y,Z,30,cmap=mpl.cm.winter,alpha=0.4)
#h = ax.plot_surface(X,Y,Z,cmap=mpl.cm.winter,alpha=0.4)
h = ax.plot_wireframe(X,Y,Z,cmap=mpl.cm.winter,alpha=0.4)
plt.colorbar(h)
plt.show()
```
