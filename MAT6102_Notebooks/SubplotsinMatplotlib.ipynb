```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
fig, axes = plt.subplots(1,3, figsize=(9,3), sharey=True)
axes[0].plot(x, y1, lw = 2)
axes[0].set_title('Sin(x) plot')
axes[0].set_xlabel('x-range')
axes[0].set_ylabel('y-range')
axes[1].plot(x, y2, lw = 2)
axes[1].set_title('Cos(x) plot')
axes[2].plot(x, y1*y2, lw = 2)
axes[2].set_title('Sin(x)Cos(x) plot')
plt.show()
```
