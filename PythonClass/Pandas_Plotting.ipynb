### Plotting in Pandas

For working with the below code your require to download the data file WQ.xlsx from https://bsrvp.github.io/PythonClass/WQ.xlsx and save it in the same directory where is your Jupyter Notebook is saved.

````python
import pandas as pd
import numpy as np
````

````python
WQData = pd.read_excel('WQ.xlsx')
````

**Note:** You can directly read the excel from the web URL (if you are connected to net). For reading the data from a web URL you need to use the following command:
