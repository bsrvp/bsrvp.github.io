```python
xl = pd.ExcelFile('WQ.xlsx')
```

```python
sheets = xl.sheet_names
print(sheets)
```

```python
WQ = pd.DataFrame()
for name in sheets:
    temp = pd.read_excel('WQ.xlsx',sheet_name=name)
    Month = pd.DataFrame([name]*temp.shape[0],columns=['Month'])
    temp = pd.concat([Month,temp],axis=1)
    #print(temp)
    if WQ.empty == True:
        WQ = temp
    else:
        WQ = pd.concat([WQ, temp])
```
