```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.subplot(3,3,1)
plt.title('Plot at location (3,3,1)')
plt.subplot(3,3,(5,9))
plt.title("Combined plot of 5 to 9")
plt.subplot(3,3,3)
plt.xlabel('X-axis (P-3)')
plt.subplots_adjust(hspace=0.85,wspace=0.5)
plt.suptitle('Demonstration of subplots')
plt.show()
```

```python
fig, ax = plt.subplots(3,3)
ax[1,0].remove()
ax[1,1].remove()
gs = ax[0,0].get_gridspec()
fig.add_subplot(gs[1,0:2])
ax[1,2].remove()
ax[2,2].remove()
fig.add_subplot(gs[1:,2])
plt.show()
```
