```python
from matplotlib import pyplot as plt
import numpy as np
x = np.arange(0, np.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title("Sine wave")
ax.set_xlabel('theta')
ax.set_ylabel('Sin(theta)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
x = np.arange(1,11)
a1.plot(x,np.exp(x))
a1.set_ylabel('exp')
a2 = a1.twinx()
a2.plot(x, np.log(x),'ro-')
a2.set_ylabel('log')
fig.legend(labels = ('exp','log'),loc='center')
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(12,6))
ax = plt.axes(projection='3d')
im = ax.contour3D(X, Y, Z, 100, cmap='ocean')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.colorbar(im)
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(12,6))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface')
plt.show()
```

```python
#%matplotlib inline
%matplotlib Qt5
import matplotlib as mpl
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

#mpl.use('qt5agg')

def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 300)
y = np.linspace(-6, 6, 300)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(12,6))
ax = plt.axes(projection='3d')
im = ax.plot_surface(X, Y, Z, cmap='ocean')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.colorbar(im)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,15,100)
y1 = x**2
y2 = np.exp(x)

plt.figure(figsize=(12,6))
plt.subplot(211)
plt.plot(x,y1,'-b')
plt.title('$x^2$')
plt.subplot(212)
plt.plot(x,y2,'-g')
plt.title('$\exp(x)$')
#plt.suptitle('Plots of $x^2$ and $\exp(x)$')
plt.show()
```
