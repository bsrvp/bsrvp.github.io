```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = pd.Series([9099827, 8615012, 278228, 2237012], name="Population", index=["Delhi","Mumbai","Vellore","Bengaluru"])
```

```python
s.median(), s.mean(), s.std()
```

```python
s.min(), s.max()
```

```python
s.quantile(q=0.25), s.quantile(q=0.5), s.quantile(q=0.75)
```

```python
s.describe()
```

```python
fig, axes = plt.subplots(1,4, figsize=(12,3))
s.plot(ax=axes[0],kind='barh',title='bar plot')
s.plot(ax=axes[1],kind='line',title='line plot')
s.plot(ax=axes[2],kind='box',title='box plot')
s.plot(ax=axes[3],kind='pie',title='pie chart')
plt.show()
```
