```python
import numpy as np
A = np.array([[17,24,1,8,15],[23,5,7,14,16],[4,6,13,20,22],[10,12,19,21,3],[11,18,25,2,9]])
rowsum = A.sum(axis=0)
print(rowsum)
colsum = A.sum(axis=1)
print(colsum)
rowsummin = rowsum.min()
print(rowsummin)
rowsummax = rowsum.max()
print(rowsummax)
colsummin = colsum.min()
print(colsummin)
colsummax = colsum.max()
print(colsummax)
diag = np.diag(A)
diagsum = diag.sum()
print(diagsum)
rdiag = np.diag(np.fliplr(A))
rdiagsum = rdiag.sum()
print(rdiagsum)
```

```python
import numpy as np
from scipy.integrate import quad, dblquad, tplquad

#First Integral
f = lambda x: np.exp(-x**2)
(val, err) = quad(f,0,5)
print(f'The value of the single integral is: {val}')

#Second Integral
g = lambda x,y: np.sin(x+y)
(val, err) = dblquad(g,0,5,lambda x: 1,lambda x:2)
print(f'The value of the double integral is: {val}')

#Third Integral
h = lambda x,y,z : np.sin(x+y)*z
(val, err) = tplquad(h,0,1,lambda x:1, lambda x:2, lambda x,y: 2, lambda x,y:3)
print(f'The value of the singel integral is: {val}')
```

```python
import numpy as np
import scipy.integrate as si
import matplotlib.pyplot as plt

x = np.linspace(0,2,15)
y = np.sqrt(x)
xi =np.linspace(0,2,100)
plt.plot(xi,np.sqrt(xi),'-r',x,y,'o')
plt.legend(['Function Plot','Data Points'])
plt.xlabel('X-values')
plt.ylabel('Function Plot')
plt.show()

int1 = si.trapz(y,x)
print(f'The integration of f(x) over the interval [0,2] using Trapezoidal rule is {int1}')

int2 = si.simps(y,x)
print(f'The integration of f(x) over the interval [0,2] using Simpsons rule is {int1}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.random.randn(100)
x2 = np.random.randn(100)

plt.subplot(221)
plt.scatter(x1,x2)
plt.title('Uncorrelated')

plt.subplot(222)
plt.scatter(x1,x1+x2)
plt.title('Weakly Positively Correlated')

plt.subplot(223)
plt.scatter(x1,-x1+x2)
plt.title('Weakly Negatively Correlated')

plt.subplot(224)
plt.scatter(x1,x1+ 0.15*x2)
plt.title('Strongly Correlated')

plt.subplots_adjust(left=0.1,right=0.95,bottom=0.1, top=0.95, wspace=0.3, hspace=0.4)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)

y1 = np.sqrt((np.exp(x)+np.sinh(x))*x/3)
y2 = 0.5*x/(2+np.cosh(x))

fig, ax = plt.subplots(2,1,figsize=(8,5))
ax[0].plot(x,y1,label='Plot of f(x)')
ax[1].plot(x,y2,label='Plot of g(x)')
for i in range(2):
    ax[i].set_xlabel('X-axis')
    ax[i].set_ylabel('Plot of the function')
    ax[i].legend()
plt.show()
fig.savefig('MyPlot.pdf')
```
