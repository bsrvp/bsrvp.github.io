```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi,np.pi,200)
y = np.sin(2*x)

plt.figure()
plt.plot(x,y, color='red', alpha = 1.0)
plt.fill_between(x,0,y,color='blue',alpha=0.25) # Here 0 after y values of the firsst curve. In otherwords, this gives the base line for filling. You can experiment with various values such as 0.5,-0.5,1,-1
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi,np.pi,200)
y = np.sin(2*x)

plt.figure()
plt.plot(x,y, color='red', alpha = 1.0)
plt.fill_between(x,0,y,y>0, color='green',alpha=0.25)
plt.fill_between(x,0,y,y<0, color='brown',alpha=0.25)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline

# We below generate some random data by generating some noise using random.randn() function
x = np.linspace(-3, 3, 50)
y = np.exp(-x**2) + 0.1 * np.random.randn(50)

# Plotting the generated data
plt.plot(x, y, 'ro', ms = 5)

# Constructing the Spline Interpolation Curve
spl = UnivariateSpline(x, y) 

# Plotting the Spline interpolation curve
xs = np.linspace(-3, 3, 1000)
plt.plot(xs, spl(xs), 'g', lw = 3)

'''
Observe that the above interpolation curve is not accurate as it is not passing through 
the given data points.  We can adjust the fitting accuracy by using set_smoothing_factor(n) 
where n ranges from 0 to 1. The more the value of n the more smoothing the curve becomes.  
You can experiemnt with various values to observe the trend
'''
spl.set_smoothing_factor(0.5)
plt.plot(xs, spl(xs), 'b', lw = 3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi,np.pi,200)
y1 = np.sin(2*x)
y2 = np.cos(2*x)

plt.figure()
plt.plot(x,y1,x,y2)
plt.fill_between(x,y1,y2,color='green',alpha=0.25)
plt.show()
```
